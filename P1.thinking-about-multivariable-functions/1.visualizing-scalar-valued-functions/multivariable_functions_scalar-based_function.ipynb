```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
```

# Scalar-based Function 

### 1. $f(x,y)=x+y$

```python
X = np.arange(-3, 3, 0.1)  
Y = np.arange(-3, 3, 0.1)  
X, Y = np.meshgrid(X, Y)    # x-y   
Z = X + Y

fig = plt.figure(figsize=(12, 8))  
ax = fig.add_subplot(111, projection='3d')  
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))  
ax.set_title("$f(x,y)=x+y$", fontsize=18)
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", fontsize=18)
ax.set_zlabel("$z$", fontsize=18)
plt.show()
```

### $f(x,y)=x^2+y^2$

```python
X = np.arange(-3, 3, 0.1)  
Y = np.arange(-3, 3, 0.1)  
X, Y = np.meshgrid(X, Y) 
Z = (X ** 2 + Y ** 2)  

fig = plt.figure(figsize=(12, 8))  
ax = fig.add_subplot(111, projection='3d')  
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))  
ax.set_title("$f(x,y)=x^2+y^2$", fontsize=18)
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", fontsize=18)
ax.set_zlabel("$z$", fontsize=18)
plt.show()  
```

```python
X = np.arange(-3, 3, 0.1)  
Y = np.arange(-3, 3, 0.1)  
X, Y = np.meshgrid(X, Y)
Z = (X ** 2 + Y ** 2)  

fig = plt.figure(figsize=(8, 8))  
ax = fig.add_subplot(111)  
ax.contourf(X, Y , Z, cmap='rainbow')
ax.set_title("$f(x,y)=x^2+y^2$", fontsize=18)
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", rotation=0, fontsize=18)
ax.set_xlim(-3.5, 3.5)
ax.set_ylim(-3.5, 3.5)
plt.show()  
```

### 2. $f(x,y)=sin(x+y)$

```python
X = np.arange(-np.pi, np.pi, 0.1)  
Y = np.arange(-np.pi, np.pi, 0.1)  
X, Y = np.meshgrid(X, Y)
Z = np.sin(X+Y)

fig = plt.figure(figsize=(12, 8))  
ax = fig.add_subplot(111, projection='3d')  
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))
ax.set_title("$f(x,y)=sin(x+y)$", fontsize=18)
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", fontsize=18)
ax.set_zlabel("$z$", fontsize=18)
plt.show()  
```

### 3. $f(x,y)=cos(x)sin(y)$

```python
X = np.arange(-np.pi, np.pi, 0.1)  
Y = np.arange(-np.pi, np.pi, 0.1)  
X, Y = np.meshgrid(X, Y)   
Z = np.sin(X) * np.cos(Y)

fig = plt.figure(figsize=(12, 8))  
ax = fig.add_subplot(111, projection='3d')  
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))  
ax.set_title("$f(x,y)=cos(x)sin(y)$", fontsize=18)
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", fontsize=18)
ax.set_zlabel("$z$", fontsize=18)
plt.show()  
```

### 4. $f(x,y)=e^{(x+y)}$

```python
X = np.arange(-2, 2, 0.1)  
Y = np.arange(-2, 2, 0.1)  
X, Y = np.meshgrid(X, Y)    
Z = np.exp(X + Y)

fig = plt.figure(figsize=(12, 8))  
ax = fig.add_subplot(111, projection='3d')  
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))  
ax.set_title("$f(x,y)=e^{(x+y)}$", fontsize=18)
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", fontsize=18)
ax.set_zlabel("$z$", fontsize=18)
plt.show()  
```
