```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
```

## Vector Fields

### 1. $f(x,y)=\bigg[\begin{matrix} x \\ y \end{matrix}\bigg]$

```python
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
X1 = X
X2 = Y

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
ax.quiver(X, Y, X1, X2, edgecolor='b', facecolor='None', linewidth=.5, cmap='autumn')
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", rotation=0, fontsize=18)
ax.set_title("$f(x,y)=[x \\ y]$", fontsize=18)
plt.show()
```

### 2. $f(x,y)=\bigg[\begin{matrix}y^3 - 9y \\ x^3-9x \end{matrix}\bigg]$ 

```python
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
X1 = Y**3 - 9*Y
X2 = X**3 - 9*X

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
ax.quiver(X, Y, X1, X2, edgecolor='b', facecolor='None', linewidth=.5, cmap='autumn')
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", rotation=0, fontsize=18)
ax.set_title("$f(x,y)=[y^3 - 9y , x^3 - 9x]$", fontsize=18)
plt.show()
```

### 3. $f(x,y,z)=\bigg[\begin{matrix}x \\ y \\ z\end{matrix}\bigg]$

```python
X = np.arange(-5, 5, 1)
Y = np.arange(-5, 5, 1)
Z = np.arange(-5, 5, 1)
X, Y, Z = np.meshgrid(X, Y, Z)
X1 = X
X2 = Y
X3 = Z

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection="3d")
ax.quiver(X, Y, Z, X1, X2, X3, length=0.1)
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", fontsize=18)
ax.set_zlabel("$z$", fontsize=18)
ax.set_title("$f(x,y,z)=[x , y , z]$", fontsize=18)
plt.show()
```

### 4. $f(x,y,z)=\bigg[\begin{matrix}yz \\ xz \\ xy\end{matrix}\bigg]$

```python
X = np.arange(-5, 5, 1)
Y = np.arange(-5, 5, 1)
Z = np.arange(-5, 5, 1)
X, Y, Z = np.meshgrid(X, Y, Z)
X1 = Y * Z
X2 = X * Z
X3 = X * Y

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection="3d")
ax.quiver(X, Y, Z, X1, X2, X3, length=0.05)
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", fontsize=18)
ax.set_zlabel("$z$", fontsize=18)
ax.set_title("$f(x,y,z)=[yz , xz , xy]$", fontsize=18)
plt.show()
```
