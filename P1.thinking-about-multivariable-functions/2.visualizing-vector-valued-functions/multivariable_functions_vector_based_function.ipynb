```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
```

# Vector-based Function 

## Parametric Curves

### 1. $f(t)=\bigg[\begin{matrix} cos(t) \\ tsin(t)\end{matrix}\bigg]$

```python
from IPython.display import YouTubeVideo
YouTubeVideo(id="bb4bSCjlFAw", width=600, height=400)
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
t = np.arange(0, 4*np.pi, 0.1)
X = np.cos(t)
Y = t * np.sin(t)

ax.set_title("$f(t)=[cos(t) \ tsin(t)]$", fontsize=18)
ax.plot(X, Y, color='r')
ax.set_xlabel("$x_1$", fontsize=18)
ax.set_ylabel("$x_2$", fontsize=18)
ax.axis('equal')
plt.show()
```

### 2. $f(t) = \bigg[\begin{matrix} t cos(t) \\ t sin(t)\end{matrix}\bigg]$

```python
theta = np.arange(-2*np.pi, 2*np.pi, 0.1)
X = theta * np.cos(theta)
Y = theta * np.sin(theta)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
ax.set_title("$f(t)=[tcos(t) , tsin(t)]$", fontsize=18)
ax.plot(X, Y, color='r')
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", rotation=0, fontsize=18)
ax.axis('equal')
plt.show()
```

## Parametric Surfaces

### $f(t,s)=\bigg[\begin{matrix}3cos(t)+cost(t)cos(s) \\\    3sin(t)+sin(t)cos(s) \\\ sin(s)\end{matrix}\bigg]$

```python
YouTubeVideo(id="345SnWfahhY", width=600, height=400)
```

```python
t = np.linspace(-np.pi, np.pi, 100)
s = np.linspace(-np.pi, np.pi, 100)
t, s =np.meshgrid(t, s)
X = 3 * np.cos(t) + np.cos(t) * np.cos(s)
Y = 3 * np.sin(t) + np.sin(t) * np.cos(s)
Z = np.sin(s)

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection="3d")
ax.set_title("$f(t,s)=[3cos(t)+cost(t)cos(s)$ , $3sin(t)+sin(t)cos(s)$ , $sin(s)]$", fontsize=18)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))
ax.set_xlabel("$x_1$", fontsize=18)
ax.set_ylabel("$x_2$", fontsize=18)
ax.set_zlabel("$x_3$", fontsize=18)
ax.set_zlim(-2, 2)
plt.show()
```

```python
import ipyvolume.pylab as plt3d

t = np.linspace(-np.pi, np.pi, 100)
s = np.linspace(-np.pi, np.pi, 100)
t, s =np.meshgrid(t, s)
X = 3 * np.cos(t) + np.cos(t) * np.cos(s)
Y = 3 * np.sin(t) + np.sin(t) * np.cos(s)
Z = np.sin(s)

plt3d.figure(screen_capture_enabled=True)
plt3d.plot_surface(X, Z, Y, color="red")
plt3d.zlabel("y")
plt3d.ylabel("z")
plt3d.ylim(-2, 2)
plt3d.show()
```
