## Partial derivatives of vector fields

### $\vec V(x,y)=\bigg [\begin{matrix} xy \\ y^2-x^2 \end{matrix} \bigg]$

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-10, 10, 10)
Y = np.linspace(-10, 10, 10)
X, Y = np.meshgrid(X, Y)
X1 = X * Y
X2 = Y**2 - X**2

plt.figure(figsize=(8, 8))
plt.quiver(X, Y, X1, X2, edgecolor='b', facecolor='None', linewidth=1, cmap='autumn')
plt.xlabel("$x$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.show()
```

### The Partial Derivative

```python
import sympy
sympy.init_printing()

x, y = sympy.symbols("x, y")
v = sympy.Matrix([[x * y], [y**2 - x**2]])
```

```python
v.diff(x)
```

```python
v.diff(y)
```

```python
partial_x1 = X
partial_x2 = -2 * X

partial_y1 = X
partial_y2 = 2 * Y

plt.figure(figsize=(10, 10))
plt.quiver(X, Y, X1, X2, linewidth=0.5, alpha=0.3, label="$f$")
plt.quiver(X, Y, partial_x1, partial_x2, edgecolor='r', facecolor='None', 
          linewidth=1, label="${\partial f}/{\partial x}$")
plt.quiver(X, Y, partial_y1, partial_y2, edgecolor='g', facecolor='None', 
          linewidth=1, label="${\partial f}/{\partial y}$")
plt.xlabel("$x$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend(fontsize=14)
plt.show()
```
