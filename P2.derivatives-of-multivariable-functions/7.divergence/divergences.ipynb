```python
import sympy
sympy.init_printing()
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
```

## Divergence Formula

```python
from IPython.display import YouTubeVideo
YouTubeVideo(id="TKlpZ0UUJTQ", width=600, height=400)
```

```python
x, y, z = sympy.symbols("x, y, z")
f = sympy.Function("f")(x, y, z)
f_grad = [f.diff(x_) for x_ in (x, y, z)]
sympy.Matrix(f_grad)
```

```python
divergence = 0
for grad in f_grad:
    divergence += grad

divergence
```

## Examples

### 1. $f(x,y)=x^2y$

```python
x, y = sympy.symbols("x, y")
f = x**2 * y
v = sympy.Matrix([-1, 2])
f_grad = [f.diff(x_) for x_ in (x, y)]
f_grad_mat = sympy.Matrix(f_grad)
f_direct = f_grad_mat.dot(v)
f_direct
```

### 2. $\vec V(x,y)=\bigg[\begin{matrix}xy \\ y^2-x^2\end{matrix}\bigg]$

```python
YouTubeVideo(id="AlXVrAOls-8", width=600, height=400)
```

```python
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
P = X * Y
Q = Y**2 - X**2

plt.figure(figsize=(8, 8))
plt.quiver(X, Y, P, Q, edgecolor='b', facecolor='None', linewidth=.5, cmap='autumn')
plt.xlabel("$x$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.title("$f(x,y)=[xy \\ y^2-x^2]$", fontsize=18)
plt.show()
```
