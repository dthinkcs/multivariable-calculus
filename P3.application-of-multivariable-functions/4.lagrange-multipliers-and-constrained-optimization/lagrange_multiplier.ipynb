## The Lagrangian

### Maximize $f(x,y)=x^2e^yy$ subject to $g(x,y)=x^2+y^2=b$

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy
sympy.init_printing()
```

```python
x, y, b =sympy.symbols("x, y, b")
f = x**2 * sympy.exp(y) * y
f_grad = [f.diff(i) for i in (x,y)]

g = x**2 + y**2
g_grad = [g.diff(i) for i in (x,y)]
lamb = sympy.Symbol("lambda")

lagrangian = f - lamb * (g - b)
lagrangian
```

```python
lagrangian_grad = [lagrangian.diff(i) for i in (x, y, lamb)]
res = sympy.solve(lagrangian_grad, (x, y, lamb)) # it will take longer because the result concludes complex imaginary roots
print("The real root:{},{} ".format(res[0], res[2]))
```

verify $b=4$

```python
from mpl_toolkits.mplot3d import Axes3D

b = 4
X = np.arange(-3, 3, 0.1)  
Y = np.arange(-3, 3, 0.1)  
X, Y = np.meshgrid(X, Y)
Z = X**2 * np.exp(Y) * Y

X0 = np.linspace(-2, 2, 100)
Y0 = np.sqrt(b - X0**2)
Z0 = X0**2 * np.exp(Y0) * Y0

fig = plt.figure(figsize=(12, 8))  
ax = fig.add_subplot(111, projection='3d')  
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'), alpha=0.6)  
ax.plot(X0, Y0, Z0, color='r')
ax.plot(X0, -Y0, Z0, color='r')
ax.scatter3D(0., np.sqrt(b), 0., c='black', s=100, marker='^')
ax.scatter3D(0., -np.sqrt(b), 0., c='black', s=100, marker='^')
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", fontsize=18)
ax.set_zlabel("$z$", fontsize=18)
plt.show()
```
