### 1. $f(x, y)=xy$, c: $x=cost$, $y=sint$, $0\leq t \leq \frac {\pi} 2$

```python
from IPython.display import YouTubeVideo
YouTubeVideo(id="uXjQ8yc9Pdg", width=600, height=400)
```

```python
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(12, 8))  
ax = fig.add_subplot(111, projection='3d')  

t = np.linspace(0, np.pi / 1.9 , 100)
X = np.cos(t)
Y = np.sin(t)
X, Y = np.meshgrid(X, Y)
Z = X * Y

t0 = np.linspace(0, np.pi / 2, 100)
X0 = np.cos(t0)
Y0 = np.sin(t0)
Z0 = X0 * Y0

ax.plot_surface(X, Y, Z, color="g", alpha=0.5)
ax.plot3D(X0, Y0, 0, color="b")
ax.plot3D(X0, Y0, Z0, color="b")
edge1 = np.array(list(zip(X0, Y0, Z0)))
edge2 = np.array(list(zip(X0, Y0, np.zeros(len(X0)))))
vtx = np.vstack([edge1, edge2])
tri = Poly3DCollection([vtx[vtx[:,0].argsort()]], alpha=0.6)
face_color = [1, 0, 0] 
tri.set_facecolor(face_color)
ax.add_collection3d(tri, zs=vtx[:,2])
ax.set_xlabel("x", fontsize=14)
ax.set_ylabel("y", fontsize=14)
ax.set_zlabel("z", fontsize=14)

ax.set_xlim(-0.2, 1)
ax.set_ylim(-0.2, 1)
ax.view_init(azim=-15)
plt.show()
```

```python
t = sympy.Symbol("t")
x = sympy.cos(t)
y = sympy.sin(t)
f = x * y
path = sympy.sqrt(x**2 + y**2).simplify()
sympy.integrate(f, (t, 0, sympy.pi / 2)) * path
```

### 2. $f(x, y)=x + y^2$, c: $x=2cost$, $y=2sint$, $0\leq t \leq \frac {\pi} 2$

```python
fig = plt.figure(figsize=(12, 8))  
ax = fig.add_subplot(111, projection='3d')  

t = np.linspace(0, np.pi , 100)
X = np.cos(t)
Y = np.sin(t)
X, Y = np.meshgrid(X, Y)
Z = X + Y**2

t0 = np.linspace(0, np.pi / 2, 100)
X0 = np.cos(t0)
Y0 = np.sin(t0)
Z0 = X0 + Y0**2

ax.plot_surface(X, Y, Z, color="g", alpha=0.3)
ax.plot3D(X0, Y0, 0, color="g")
ax.plot3D(X0, Y0, Z0, color="b")
ax.plot3D(X0, np.zeros(len(X0)), np.zeros(len(X0)), color="b")
ax.plot3D(X0, np.zeros(len(X0)), X0 , color="b")
ax.plot3D(np.zeros(len(X0)), Y0, np.zeros(len(X0)), color="b")
ax.plot3D(np.zeros(len(X0)), Y0, Y0**2, color="b")

edge1 = np.array(list(zip(X0, Y0, Z0)))
edge2 = np.array(list(zip(X0, Y0, np.zeros(len(X0)))))
edge3 = np.array(list(zip(X0, np.zeros(len(X0)), np.zeros(len(X0)))))
edge4 = np.array(list(zip(X0, np.zeros(len(X0)), X0)))
edge5 = np.array(list(zip(np.zeros(len(X0)), Y0, np.zeros(len(X0)))))
edge6 = np.array(list(zip(np.zeros(len(X0)), Y0, Y0**2)))

vtx = np.vstack([edge1, edge2, edge3, edge4, edge5, edge6])
tri = Poly3DCollection([vtx[vtx[:,0].argsort()]], alpha=0.6)
face_color = [1, 0, 0] 
tri.set_facecolor(face_color)
ax.add_collection3d(tri, zs=vtx[:,2])
ax.set_xlabel("x", fontsize=14)
ax.set_ylabel("y", fontsize=14)
ax.set_zlabel("z", fontsize=14)

ax.set_xlabel("x", fontsize=14)
ax.set_ylabel("y", fontsize=14)
ax.set_zlabel("z", fontsize=14)
plt.show()
```

```python
t = sympy.Symbol("t")
x = 2 * sympy.cos(t)
y = 2 * sympy.sin(t)
f = x + y**2
c1 = sympy.sqrt(x**2 + y**2).simplify()
integral_c1 = sympy.integrate(f, (t, 0, sympy.pi / 2)) * c1
```

```python
x = 0
y = 2 - t
f = x + y**2
c2 = sympy.sqrt(0 + y.diff("t")**2)
integral_c2 = sympy.integrate(f, (t, 0, 2)) * c2
```

```python
x = t
y = 0
f = x + y**2
c3 = sympy.sqrt(x.diff("t")**2 + 0)
integral_c3 = sympy.integrate(f, (t, 0, 2)) * c3
```

```python
integral = integral_c1 + integral_c2 + integral_c3
integral
```
